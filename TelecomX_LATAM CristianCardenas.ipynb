Datos vacÃ­os (ya tratados): 224 (~3%)

Factores que influyen en el churn
1. Tipo de contrato

- Month-to-month: Muy alta (~43%)
- One year: Baja (~11%)
- Two year: Muy baja (~3%)

ConclusiÃ³n: Los contratos mensuales permiten irse fÃ¡cilmente. Los contratos largos reducen dramÃ¡ticamente la evasiÃ³n, probablemente por compromiso, penalizaciones o descuentos.

2. MÃ©todo de pago

- Electronic check: Alta (~45%)
- Mailed check: Media (~19%)
- Credit card (automatic): Baja (~15%)
- Bank transfer (automatic): Baja (~14%)

ConclusiÃ³n: El pago automÃ¡tico fomenta la permanencia, mientras que el cheque electrÃ³nico se asocia a desorganizaciÃ³n o menor compromiso financiero.

3. GÃ©nero

- Hombre: ~26.2%
- Mujer: ~26.3%

ConclusiÃ³n: No hay diferencia significativa. El gÃ©nero no influye en la tasa de abandono.

4. Tipo de Internet

- Fiber optic: Alta (~42%)
- DSL: Media (~19%)
- No Internet: Muy baja (~7%)

ConclusiÃ³n: La fibra Ã³ptica presenta la mayor tasa de churn, posiblemente por problemas tÃ©cnicos, costos elevados o expectativas no cumplidas.

Otras observaciones clave

- La modalidad de contrato mensual es el mayor predictor de abandono.
- Los clientes nuevos (tenure bajo) tienen mÃ¡s probabilidades de irse.
- Las tarifas mÃ¡s bajas (menos de $30) tienen tasas de churn altas.
- El cheque electrÃ³nico aparece como factor de riesgo en mÃºltiples anÃ¡lisis.

Recomendaciones para la empresa Telecom

1. Fomentar contratos a largo plazo con beneficios atractivos.
2. Incentivar pagos automÃ¡ticos.
3. Revisar la calidad y precios del servicio de fibra Ã³ptica.
4. DiseÃ±ar campaÃ±as de retenciÃ³n para nuevos clientes.
